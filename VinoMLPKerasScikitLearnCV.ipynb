```python
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
import tensorflow as tf
from tensorflow.keras import layers, models, wrappers
```

```python
datos=pd.read_csv("Vino.csv", header = None)
x = datos.iloc[:, 1:14].values
y = datos.iloc[:, 0].values
```

```python
#escalar los datos para que tengan media = 0 y desviacion estandar = 1
scaler = StandardScaler().fit(x)
x = scaler.transform(x)
```

```python
#etiquetas con valores de 0 a nClases-1
le = LabelEncoder().fit(y)
y = le.transform(y)
```

```python
def modelo():
    model = models.Sequential()
    model.add(layers.Dense(100, input_shape=(13,), activation='relu')) #neuronas ocultas, variables de entrada, f(x) de activacion
    model.add(layers.Dense(3, activation='softmax')) #clases, f(x) de activacion
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
clf = wrappers.scikit_learn.KerasClassifier(build_fn = modelo, epochs = 200, batch_size = 128)
```

```python
scores = cross_val_score(clf, x, y, cv = 5)
```

```python
print(scores)
print(scores.mean())
```
